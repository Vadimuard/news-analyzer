### Generate the data <a class="anchor" id="get_data"></a>
First, we need to receive the data from newsapi.org

```python
# import all the packages required for this project
import calendar
from itertools import chain

import numpy as np
import requests
import gensim

import pandas as pd
import pymongo as pymongo
from adjustText import adjust_text
from bson import ObjectId
from dotenv import dotenv_values
from matplotlib import pyplot as plt
import seaborn as sns
from nltk import download, word_tokenize
from nltk.corpus import stopwords
from scipy.ndimage import gaussian_filter1d
from sklearn.manifold import TSNE
```

```python
# get credentials from .env file

config = dotenv_values(".env")

sources_arr = [['bbc-news', 'bloomberg', 'msnbc'],
               ['the-hill', 'the-wall-street-journal', 'vice-news', ],
               ['the-washington-times', 'the-huffington-post', 'business-insider']]


def make_request(sources):
    url = f"https://newsapi.org/v2/everything?sources={sources}&from=2021-05-10&sortBy=popularity&pageSize=50&apiKey={config['NEWS_API_TOKEN']}"
    resp_articles = requests.get(url)
    articles = resp_articles.json()['articles']
    return articles, resp_articles.json

all_articles = []
for sources in sources_arr:
    for source in sources:
        articles, res = make_request(source)
        all_articles.extend(articles)
print(f"number of articles received: {len(all_articles)}")
print("First 5 articles to understand the data structure:")
print(all_articles[:5])
```

In this step we need to initialize our DataBase. We'll use MongoDB here

```python
client = pymongo.MongoClient(
    f"mongodb+srv://admin:{config['MONGO_PASS']}@cluster0.mfqgk.mongodb.net/newsdb?retryWrites=true&w=majority")
db = client.newsdb
articles_collection = db["articles"]
```

Finally, we can insert articles into the database

```python
for article in all_articles:
    articles_collection.insert_one(article)
```

### Validate the data <a class="anchor" id="validate"></a>
In this step we remove all the duplicate data

```python
all_articles = articles_collection.find({}, {"_id": 1, "title": 1})
articles_titles = set()
print(f"Number of documents in the collection before validation: {articles_collection.count_documents({})}")
for article in all_articles:
    if article['title'] not in articles_titles:
        articles_titles.add(article['title'])
    else:
        articles_collection.delete_one({'_id': ObjectId(article['_id'])})

print(f"Number of documents in the collection after validation: {len(articles_titles)}")
```

### Process the data <a class="anchor" id="process"></a>

```python
import gc
i = 0
art = articles_collection.find({}, {"_id": 1, "description": 1})
description_list = [doc['description'] for doc in art if doc['description'] is not None]
big_desc_str = ' '.join(description_list)

download('stopwords')
download('punkt')

st_wrd = set(stopwords.words('english'))
tokens = word_tokenize(big_desc_str)
calldwr = [wd.lower() for wd in list(calendar.day_name)]
words = [word.lower() for word in tokens
         if word.isalpha()
         and word.lower() not in st_wrd
         and len(word) >= 3
         and word.lower() not in calldwr]
wordfreq = {}
for w in words:
    if w not in wordfreq:
        wordfreq.update({w: words.count(w)})

wordfreq_sorted = sorted(wordfreq.items(), key=lambda item: item[1], reverse=True)

sorted_dict = {w[0]: w[1] for w in wordfreq_sorted}
sorted_df = pd.DataFrame.from_dict(sorted_dict, orient='index')
sorted_df.to_pickle('data/wordcount_data.pickle')

model = gensim.models.KeyedVectors.load_word2vec_format('models/GoogleNews-vectors-negative300.bin', binary=True)
vector_list = [model[word] for word in words if word in model.key_to_index]
# Create a list of the words corresponding to these vectors
words_filtered = [word for word in words if word in model.key_to_index]
# Zip the words together with their vector representations

word_vec_zip = zip(words_filtered, vector_list)
# Cast to a dict so we can turn it into a DataFrame
word_vec_dict = dict(word_vec_zip)
df = pd.DataFrame.from_dict(word_vec_dict, orient='index')
df.to_pickle('data/dataframe.pickle')
```

```python
# collect garbage
del model
del vector_list
del words_filtered
del word_vec_zip
del word_vec_dict
del df
del sorted_df
del sorted_dict
gc.collect()
```

### Analyze the data to determine the most commonly used words in the dataset <a class="anchor" id="mostcommon"></a>
Also, plot the correlations between words in the dataset

```python
vector_df = pd.read_pickle('data/dataframe.pickle')
wordcount_df = pd.read_pickle('data/wordcount_data.pickle')
# Initialize t-SNE
tsne = TSNE(n_components=2, init='random', random_state=10, perplexity=300)

analyzed_data = vector_df[:3000]
tsne_df = tsne.fit_transform(analyzed_data)
words = analyzed_data.index.values
named_plot_data = {words[i]: [tsne_df[i, 0], tsne_df[i, 1]] for i in range(len(words))}

sns.set()
# Initialize figure
fig, ax = plt.subplots(figsize=(12, 9))
sns.scatterplot(tsne_df[:, 0], tsne_df[:, 1], alpha=0.5)

texts = []
words_to_plot = wordcount_df[:150].index.values

for word in words_to_plot:
    if word in named_plot_data.keys():
        val = named_plot_data[word]
        texts.append(plt.text(named_plot_data[word][0], named_plot_data[word][1], word, fontsize=8))

# Plot text using adjust_text (because overlapping text is hard to read)
adjust_text(texts, expand_points=(1, 1), expand_text=(1, 2),
            arrowprops=dict(arrowstyle="-", color='black', lw=1))
plt.savefig('media/words_corr.png')
```

Plot the frequency each word happens to be in the title of the article

```python
words_to_display = list(words_to_plot[:40])
data_of_words_to_display = [wordcount_df.loc[w, 0] for w in words_to_display]
plt.figure(figsize=(8,9))
plt.xticks(range(len(words_to_display)), words_to_display, rotation='vertical')
plt.bar(range(len(data_of_words_to_display)), data_of_words_to_display, align='center')
plt.savefig('media/words_freq.png')
```

### Analyze the article frequency for each source <a class="anchor" id="time_articles"></a>
Plot the number of the articles for each source depending on time

```python
sources_arr = list(chain(*sources_arr))
result_dictionary = {}
for source in sources_arr:
    result_dictionary[source] = {'aggr': 0, 'arr': []}

date = 1
days = 25
for i in range(days):
    new_date = date + i
    for hrs in [0, 6, 12, 18]:
        hour = hrs
        new_hour = hrs + 6
        if new_hour == 24:
            new_new_date = new_date + 1
            new_hour = 00
        else:
            new_new_date = new_date
        odate = '2021-05-{:02d}T{:02d}:00:00'.format(new_date, hour)
        ndate = '2021-05-{:02d}T{:02d}:00:00'.format(new_new_date, new_hour)
        for source in sources_arr:
            cnt = articles_collection.count_documents({"publishedAt": {"$gt": odate,
                                                            "$lt": ndate},
                                            "source.id": source})
            result_dictionary[source]['aggr'] += cnt
            if len(result_dictionary[source]['arr']) != 0:
                result_dictionary[source]['arr'].append(result_dictionary[source]['arr'][-1] + cnt)
            else:
                result_dictionary[source]['arr'].append(cnt)

aggr = 0
for key in result_dictionary.keys():
    aggr += result_dictionary[key]['aggr']

true_aggr = articles_collection.count_documents({"publishedAt": {"$gt": "2021-05-10T00:00:00",
                                                      "$lt": "2021-05-30T00:00:00"}})

plt.figure(figsize=(16, 9))
for key in result_dictionary.keys():
    ysmoothed = gaussian_filter1d(result_dictionary[key]['arr'], sigma=1.15)
    plt.plot(ysmoothed, label=key, linewidth=2.2)

plt.legend(result_dictionary.keys())
plt.savefig('media/time_articles.png')
```

### Analyze the data to determine the most popular articles, and the most valuable words in them <a class="anchor" id="article_corr"></a>
Plot the correlations between these articles

```python
start_date = 22
articles_curs = articles_collection.find(
    {"publishedAt": {"$gt": '2021-05-{:02d}T00:00:00'.format(start_date)}}, {'description': 1})
articles = list(articles_curs)

dictionary = {}

def vec_repr_of_sentence(sentence):
    wordlist = word_tokenize(sentence)
    list_of_vectors = [vector_df.loc[word, :].values for word in wordlist if word in vector_df.index.values]
    if len(list_of_vectors) == 0:
        return None
    sentence_vec = np.mean(list_of_vectors, axis=0)
    return sentence_vec

for art in articles[:500]:
    if art['description'] is not None:
        sentence = art['description']
        vec = vec_repr_of_sentence(sentence)
        if vec is not None:
            dictionary[sentence] = []
            dictionary[sentence].extend(list(vec))

sentence_df = pd.DataFrame.from_dict(dictionary, orient='index')
tsne = TSNE(n_components=2, init='random',learning_rate=500, random_state=10, perplexity=50)
tsne_df = tsne.fit_transform(sentence_df)

sns.set()
# Initialize figure
fig, ax = plt.subplots(figsize=(12, 9))
sns.scatterplot(tsne_df[:, 0], tsne_df[:, 1], alpha=0.5)
named_plot_data = {sentence_df.index.values[i]: [tsne_df[i, 0], tsne_df[i, 1]]
                   for i in range(len(sentence_df.index.values))}
texts = []
words_to_plot = wordcount_df[:20]
for sentence in sentence_df.index.values:
    sentence_words = word_tokenize(sentence)
    pop_words_count = 0
    for word in words_to_plot.index.values:
        if word in sentence_words:
            pop_words_count += sentence_words.count(word)
    if pop_words_count >= 4:
        article_title = articles_collection.find({'description': sentence}, {'title': 1})[0]['title']
        texts.append(plt.text(named_plot_data[sentence][0], named_plot_data[sentence][1], article_title[:20] + "...", fontsize=8))

adjust_text(texts, expand_points=(1, 1), expand_text=(1, 2),
            arrowprops=dict(arrowstyle="-", color='black', lw=1))
plt.savefig('media/articles_corr.png')
```
